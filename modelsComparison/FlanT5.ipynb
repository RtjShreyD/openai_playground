```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")

inputs = tokenizer("A step by step recipe to make bolognese pasta:", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

```python
inputs = tokenizer("What is the concept of magnetism?", return_tensors="pt")
outputs = model.generate(**inputs)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))
```

Now how to optimise the output, while using T5 large model.

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large") # large
#tokenizer = T5Tokenizer.from_pretrained("t5-large")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
```

```python
%%time

my_text="How to make a cup of tea?"
inputs = tokenizer(my_text, return_tensors="pt")

outputs = model.generate(**inputs, \
  min_length=256, \
  max_new_tokens=512, \
  length_penalty = 2, \
  num_beams=4, \
  no_repeat_ngram_size=2, \
  #num_return_sequences 2,1
  early_stopping=True)

output_text_Flan_t5 = tokenizer.batch_decode(outputs, skip_special_tokens=True)

print (output_text_Flan_t5)
```

```python
%%time

my_text="What is the concept of magnetism?"
inputs = tokenizer(my_text, return_tensors="pt")

outputs = model.generate(**inputs)

output_text_Flan_t5 = tokenizer.batch_decode(outputs, skip_special_tokens=True)

print (output_text_Flan_t5)
```
